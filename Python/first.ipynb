```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 6)
```

```python
# Read in the data

df = pd.read_csv(r'data\movies.csv')
df.head()
```

```python
df.shape
```

```python
# Overview of the Dataframe

df.info()
```

```python
# Check for missing data

df.isnull().sum()
```

```python
# Check thee percentage of missing values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f'{col} - {round(pct_missing * 100, 2)}%')
```

```python
df['year'] = df['released']
```

```python
df_sort = df.sort_values(by='gross', ascending=False)
df_sort.head()
```

```python
df.duplicated().sum()
```

```python
# Gross Correlation

df_corr = df.corr(numeric_only=True, method='pearson') # pearson, kendall, spearman
df_corr
```

```python
sns.heatmap(df_corr, annot=True, cbar='')
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
plt.scatter(x=df_sort['budget'], y=df_sort['gross'], alpha=0.2)
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.title('Budget vs Gross Earnings')
plt.show()
```

```python
sns.regplot(df_sort, x='budget', y='gross', line_kws={'color': 'green'})
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.title('Budget vs Gross Earnings')
plt.show()
```

```python
df_num = df.copy()

for col in df_num.columns:
    if(df_num[col].dtype == 'object'):
        df_num[col] = df_num[col].astype('category')
        df_num[col] = df_num[col].cat.codes
df_num
```

```python
df_num_corr = df_num.corr(method='pearson')
sns.heatmap(df_num_corr, annot=True, cbar='')
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
